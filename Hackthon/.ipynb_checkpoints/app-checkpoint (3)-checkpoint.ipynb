```python
def sort_data(data):
    d= data.loc[data['Vaccinated']==1]
    a = data.loc[(data['Age']>=60)].sort_values(["Total", "Age"], ascending=[False, False])
    a = a.loc[(data['Vaccinated']==0)].sort_values(["Total", "Age"], ascending=[False, False])
    data = data.drop(a.index)
    b = data.loc[data['Age']>=40].sort_values(["Total", "Age"], ascending=[False, False])
    b = b.loc[(data['Vaccinated']==0)].sort_values(["Total", "Age"], ascending=[False, False])
    c = pd.concat([d,a,b])
    return c
```
